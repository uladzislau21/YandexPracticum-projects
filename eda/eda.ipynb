### Работа с пропусками

```python
#выведем количество пропусков в колонках и их процент от отбщего количества наблюдений
nas = pd.DataFrame(
    {
        'missing_total': flats.isna().sum(),
        'in_perc': flats.isna().mean()
    }
)

(nas[nas['missing_total'] != 0].
 sort_values('in_perc', ascending=False).
 style.format({'in_perc':'{:.2%}'})
)
```

<b>✔️ Комментарий ревьюера:</b> Молодец, что посчитал долю пропуков! Это очень важный шаг в работе. Так видим, что необходимо обработать, а чем можно пожертвовать

#### Колонка ceiling_height

Пропуски в этой колонке составляют почти **39%** от всех наблюдений. Удалять их нельзя. Необходимо заменить.

```python
#проанализируем пропуски в колонке ceiling_eight
flats.query('ceiling_height.isna()').head()
```

Причина возникновения пропусков: люди подавали объявления без высоты потолков. Типичная высота потолка 2-3 м. Есть исключения. Например в пентхаусах и многоуровневых квартирах они могут быть до [7 м](https://asninfo.ru/comfortlive/210-samyye-bolshiye-kvartiry-sankt-peterburga). В наших данных всего лишь 26 наблюдений имеют высоту потолков до 7 метров. Медиана и среднее не сильно отличаются, а экстримальных значений мало (они не сдвигают медиану и среднее сильно) поэтому, можем заменить пропущенные значения в колонке **ceiling_height** на среднее.

```python
ceil_height_mean = flats['ceiling_height'].mean()
flats['ceiling_height'].fillna(ceil_height_mean, inplace=True)
flats['ceiling_height'].isna().sum() #проверим как произошла замена
```

<div class="alert alert-success">
<b>✔️ Комментарий ревьюера:</b> Удачное заполнение пропусков

#### Колонка balcony

Пропуски сотавляют почти 49%. Здесь необходимо изучить природу пропусков. Может пропуск обозначает 0 - отсутсвие балкона.

```python
flats.query('balcony.isna()').head()
```

Можно сделать предположение, что количество балконов зависит от площади квартиры (чем больше квартира, тем больше в ней может быть балконов). Построим график, чтобы выяснить это.

```python
flats.plot(x='balcony', y='total_area', kind='scatter')
```

Из графика видно, что площадь не влияет на количество балконов. Большие квартиры могут вообще не иметь балконов. Квартиры поменьше могут иметь 5 балконов. Заменим пропуски на 0 - отсутствие балкона.

```python
flats['balcony'].fillna(0, inplace=True)
flats['balcony'].isna().sum() #проверим как произошла замена
```

Остальные поселения не имеют информации о расстоянии до центра СПБ. Их можно узнать, например, рассчитав расстояние между центрами двух городов. Используем библиотеку **geopy** и [этот туториал](https://towardsdatascience.com/geocode-with-python-161ec1e62b89), чтобы получить широту и долготу каждого поселения, а затем по ним рассчитаем расстояние до центра СПБ.

```python
#вынесем поселения без расстояния до центра в отдельный лист и добавим к названиям область для повышения точности
#определения координат
region = ', Ленинградская область'
nas_locality_name = flats.query('cityCenters_nearest.isna() and ~locality_name.isna()')['locality_name'].unique()
nas_locality_name = [str(el) + region for el in nas_locality_name]

#определим locator как в туториале
locator = gp.Nominatim(user_agent='myGeocoder')

#создадим пустой словарь куда будем вставлять название местности как ключ,
#а словарь с широтой и долготой как название ключа
coordinates = {}

#напишем цикл который проходит по названиям местностей,
#по нему находит полное название местности с помощью функции geocode,
#затем с помощью атрибутов latitude и longitude записывает широту и
#долготу в словарь
for name in nas_locality_name:
    location = locator.geocode(name)
    if location is None:
        continue
    else:
        lat, lon = location.latitude, location.longitude
        coordinates[name] = [lat, lon]
```

Циклу необходимо 2м 40 секунд, чтобы закончить работу. Наврное, можно будет оптимизировать. У нас есть словарь, хранящий название местности и ее координаты. Координаты СПБ - **(59.9386, 30.3141)**. Теперь с помощью той же библиотеки и функции **geodesic** рассчитаем расстояние между двумя точками.

```python
spb_lon_lat = (59.9386, 30.3141)
```

```python
#создадим словарь для хранения расстояний в метрах
cityCenter_distances = {}

#пройдемся циклом по словарю с координатами и высчитаем расстояние
for key, value in coordinates.items():
    distance = geodesic(coordinates[key], spb_lon_lat).m
    cityCenter_distances[key] = distance
```

```python
print(cityCenter_distances)
```

```python
#удалим область из названия ключей
cityCenter_distances = {key.replace(region, ''):value for key, value in cityCenter_distances.items()}
cityCenter_distances
```

Теперь у нас есть словарь с названием местности и расстоянием до центра СПБ в метрах. Проверка в Яндекс.Картах показала, что расстояния довольно точные. Надо добавить их в датафрэйм в колонку **cityCenters_nearest**.

<div class="alert alert-success">
<b>✔️ Комментарий ревьюера:</b> Мне очень нравится, как ты заполнил пропуски, прибегнув к помощи сторонней библиотеки! Решение настоящего аналитика!

```python
def distance_imputer(row):
    name = row['locality_name']
    if name in cityCenter_distances.keys():
        return cityCenter_distances[name]
    else:
        return row['cityCenters_nearest']
        
flats['cityCenters_nearest'] = flats.apply(distance_imputer, axis=1)
```

Были проверены следующие гипотезы:

1. **Влияет ли площадь квартиры на ее цену?**

Площадь квартиры влияет на ее цену (коэффициент корреляции - 0.76 для квартир по всему городу и 0.75 для квартир в центре) вне зависимости от ее расположения.

2. **Влияет ли количество комнат на цену квартиры?**

Количество комнат влияет на цену квартиры вне зависимости от ее расположения, но не напрямую, а через площадь квартиры, так как количество комнат зависит от алощади квартиры.

3. **Влияет ли удаленность от центра города на цену квартиры?**

Для квартир, расположенных в центре, цена никак не зависит распложения в этом центре, а вот для квартир по всему городу цена зависит от удаленности от центра (коэффициент корреляции -0.64).

4. **Влияет ли этаж на стоимость квартиры?**

Для квартир в центре и по всему городу этаж одинаково влияет на цену: квартиры на первом этаже стоят дешевле остальных, квартиры на последнем этаже стоят подороже, и квартиры на других этажах стоят дороже всех.

5. **Влияет ли день, месяц, или год размещения объявления на цену квартиры?**

День размещения объявления не влияет на цену любых квартир. Месяц размещения влияет на цену квартир, расположенных в центре (самые дешевые квартиры в июне, самые дорогие - в марте и сентябре.), но не для квартир по всему городу. Год размещенния объявления одинаково влияет на квартиры в центре и по всему городу, в период с 2014 по 2019 год квартиры дорожали.

5. **Какие населенные пункты имеют самую дорогую и самую дешевую цену за кв.м?**

Всеволожск имеет самую маленькую цену за кв.м. Санкт-Петербург имеет самую большую.
